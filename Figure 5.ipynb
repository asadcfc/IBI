```python
import numpy as np
import pandas as pd
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as metrics
```

```python
## Step 1: Calculate the genetic risk score for Random 1, GWAS Top 1, IBI Top 1, Random 3, GWAS Top 3, IBI Top 3 using the Primary Genomic files
## Step 2: Save the genetic risk score files
## Step 3: Load the secondary files of genetic risk scores 

random1_df = pd.read_csv('randomtop1.csv', index_col=0) # Random 1 #
GWAS_Top_1_df = pd.read_csv('gwastop1.csv', index_col=0) # GWAS Top 1 #
IBI_Top_1_df = pd.read_csv('ibitop1.csv', index_col=0) # IBI Top 1 #
random3_df = pd.read_csv('randomtop3.csv', index_col=0) # Random 3 #
GWAS_Top_3_df = pd.read_csv('gwastop3.csv', index_col=0) # GWAS Top 3 #
IBI_Top_3_df = pd.read_csv('ibitop3.csv', index_col=0) # IBI Top 3 #
```

```python
## Figure 5a: Random 1 ##

y_true = np.array(random1_df['hypertension'].tolist())
y_scores = np.array(random1_df['genetic_risk_scores'].tolist())
fpr, tpr, threshold = metrics.roc_curve(y_true, y_scores)
roc_auc = metrics.auc(fpr, tpr)

import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right',  fontsize=16)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel('1 - specificity', fontname="Times New Roman", size=22,fontweight="bold")
plt.ylabel('Sensitivity', fontname="Times New Roman", size=22,fontweight="bold")
plt.title('Random 1', fontname="Times New Roman", size=22,fontweight="bold")
plt.yticks(fontname="Times New Roman", size=16)
plt.xticks(fontname="Times New Roman", size=16)
plt.show()
```
